## Calculate Core KPIs

### Subtask:
Compute key performance indicators: Revenue per Category (with and without discounts), % Sales Driven by Discounts vs Full Price, Best Sellers Contribution to Revenue, and Ratings vs Sales Correlation.


**Reasoning**:
The subtask requires preparing the data for KPI calculations by copying the DataFrame, filling missing values in 'purchased_last_month', and creating 'actual_revenue' and 'potential_revenue_if_full_price' columns. Then, I will calculate Revenue per Category (with and without discounts), % Sales Driven by Discounts vs Full Price, Best Sellers Contribution to Revenue, and Ratings vs Sales Correlation as instructed.



```python
import pandas as pd
import numpy as np

# 1. Prepare data for KPI calculations:
# Create a copy of the DataFrame to perform KPI calculations without altering the original df.
# Fill missing values in the purchased_last_month column with 0
kpi_df = df.copy()
kpi_df['purchased_last_month'] = kpi_df['purchased_last_month'].fillna(0)

# Create 'actual_revenue' by multiplying 'discounted_price' and 'purchased_last_month'.
kpi_df['actual_revenue'] = kpi_df['discounted_price'] * kpi_df['purchased_last_month']

# Create 'potential_revenue_if_full_price' by multiplying 'original_price' and 'purchased_last_month'.
kpi_df['potential_revenue_if_full_price'] = kpi_df['original_price'] * kpi_df['purchased_last_month']

# Handle any resulting NaN values in revenue columns by dropping the rows.
kpi_df.dropna(subset=['actual_revenue', 'potential_revenue_if_full_price'], inplace=True)

print("Data preparation for KPI calculations completed.")
print("KPI DataFrame Info after preparation:")
kpi_df.info()

# 2. Calculate Revenue per Category (with and without discounts):
revenue_per_category = kpi_df.groupby('product_category').agg(
    actual_revenue_sum=('actual_revenue', 'sum'),
    potential_revenue_sum=('potential_revenue_if_full_price', 'sum')
).reset_index()
revenue_per_category.rename(columns={
    'actual_revenue_sum': 'Revenue with Discounts per Category',
    'potential_revenue_sum': 'Potential Revenue without Discounts per Category'
}, inplace=True)
print("\nRevenue per Category:")
print(revenue_per_category.head())

# 3. Calculate % Sales Driven by Discounts vs Full Price:
# Identify sales that occurred with a discount
total_discounted_revenue = kpi_df[kpi_df['discount_percentage'] > 0]['actual_revenue'].sum()

# Identify sales that occurred at full price
total_full_price_revenue = kpi_df[kpi_df['discount_percentage'] == 0]['actual_revenue'].sum()

# Calculate total overall revenue
total_overall_revenue = total_discounted_revenue + total_full_price_revenue

# Compute the percentage
percentage_discounted_sales = (total_discounted_revenue / total_overall_revenue) * 100 if total_overall_revenue > 0 else 0
percentage_full_price_sales = (total_full_price_revenue / total_overall_revenue) * 100 if total_overall_revenue > 0 else 0

print("\nSales Driven by Discounts vs Full Price:")
print(f"Total Discounted Revenue: {total_discounted_revenue:.2f}")
print(f"Total Full Price Revenue: {total_full_price_revenue:.2f}")
print(f"Percentage of Total Revenue from Discounted Sales: {percentage_discounted_sales:.2f}%")
print(f"Percentage of Total Revenue from Full-Price Sales: {percentage_full_price_sales:.2f}%")

# 4. Calculate Best Sellers Contribution to Revenue:
# Filter for best sellers
best_sellers_df = kpi_df[kpi_df['is_best_seller'] == 'Best Seller']

# Calculate revenue for best sellers
best_seller_revenue = best_sellers_df['actual_revenue'].sum()

# Calculate total revenue for all products
total_revenue_all_products = kpi_df['actual_revenue'].sum()

# Compute the ratio
best_seller_contribution_percentage = (best_seller_revenue / total_revenue_all_products) * 100 if total_revenue_all_products > 0 else 0

print("\nBest Sellers Contribution to Revenue:")
print(f"Best Seller Revenue: {best_seller_revenue:.2f}")
print(f"Total Revenue (All Products): {total_revenue_all_products:.2f}")
print(f"Best Sellers Contribution to Total Revenue: {best_seller_contribution_percentage:.2f}%")

# 5. Calculate Ratings vs Sales Correlation:
# Select relevant columns and drop NaNs
ratings_sales_df = kpi_df[['product_rating', 'purchased_last_month']].dropna()

# Calculate Pearson correlation coefficient
if not ratings_sales_df.empty:
    correlation = ratings_sales_df['product_rating'].corr(ratings_sales_df['purchased_last_month'])
    print("\nRatings vs Sales Correlation (Pearson):")
    print(f"Correlation between Product Rating and Purchased Last Month: {correlation:.2f}")
else:
    print("\nNot enough data to calculate Ratings vs Sales Correlation after dropping NaNs.")
```

## Analyze Dead Weight Products

### Subtask:
Identify products that receive heavy discounts but show no significant sales lift by correlating discount levels with sales volume or revenue to pinpoint underperforming discounted items.


**Reasoning**:
I need to calculate the 75th percentile of discount_percentage, filter for heavily discounted products, then calculate the 25th percentile of purchased_last_month for those products, and finally identify products that are heavily discounted but have low sales volume (dead weight products) as per the subtask instructions.



```python
import numpy as np

# 1. Calculate the 75th percentile of the 'discount_percentage' from the kpi_df DataFrame.
# Filter out NaN values before calculating percentile to ensure accurate threshold.
discount_threshold = kpi_df['discount_percentage'].quantile(0.75)
print(f"75th percentile of Discount Percentage: {discount_threshold:.2f}%")

# 2. Filter the kpi_df to create a new DataFrame, heavily_discounted_products_df,
# containing only products with a 'discount_percentage' above this 75th percentile threshold.
heavily_discounted_products_df = kpi_df[kpi_df['discount_percentage'] > discount_threshold].copy()
print(f"\nNumber of heavily discounted products: {len(heavily_discounted_products_df)}")

# 3. From heavily_discounted_products_df, calculate the 25th percentile of 'purchased_last_month'
# to identify a threshold for 'low sales volume'.
# Ensure 'purchased_last_month' has no NaNs before calculating percentile
# (though already handled by fillna(0) in kpi_df creation, good to be cautious).
sales_volume_threshold = heavily_discounted_products_df['purchased_last_month'].quantile(0.25)
print(f"25th percentile of 'Purchased Last Month' for heavily discounted products: {sales_volume_threshold:.2f}")

# 4. Further filter heavily_discounted_products_df to find products where 'purchased_last_month'
# is below this 25th percentile. These products are considered 'dead weight products'.
dead_weight_products_df = heavily_discounted_products_df[
    heavily_discounted_products_df['purchased_last_month'] < sales_volume_threshold
].copy()

# 5. Display the product_title, 'product_category', 'discount_percentage', 'purchased_last_month',
# and 'actual_revenue' for the identified 'dead weight products'.
print(f"\nTotal number of dead weight products found: {len(dead_weight_products_df)}")
print("\nDead Weight Products (heavy discounts, low sales volume):")
print(dead_weight_products_df[['product_title', 'product_category', 'discount_percentage', 'purchased_last_month', 'actual_revenue']].head())
```

## Investigate Trust Gap

### Subtask:
Analyze products with high ratings but low sales, and conversely, low ratings but high sales. This step will seek to understand potential reasons for these discrepancies, such as marketing issues or price sensitivity.


**Reasoning**:
The subtask requires calculating percentile thresholds for product ratings and sales, then filtering the kpi_df to identify products with 'high ratings but low sales' and 'low ratings but high sales'. I will perform these calculations and filtering steps within a single code block, then display the results as requested.



```python
import numpy as np

# 1. Define thresholds for 'high ratings', 'low ratings', 'high sales', and 'low sales'.
# Note: Ensure 'product_rating' and 'purchased_last_month' columns are cleaned of NaNs before calculating percentiles if not already done.
# kpi_df was created by filling NaNs in 'purchased_last_month' with 0 and dropping rows with NaN in revenue, so these columns should be suitable for percentile calculation.

# Calculate the 75th percentile of 'product_rating' to define the high rating threshold.
high_rating_threshold = kpi_df['product_rating'].quantile(0.75)
print(f"75th percentile (High Rating Threshold): {high_rating_threshold:.2f}")

# Calculate the 25th percentile of 'product_rating' to define the low rating threshold.
low_rating_threshold = kpi_df['product_rating'].quantile(0.25)
print(f"25th percentile (Low Rating Threshold): {low_rating_threshold:.2f}")

# Calculate the 75th percentile of 'purchased_last_month' to define the high sales threshold.
high_sales_threshold = kpi_df['purchased_last_month'].quantile(0.75)
print(f"75th percentile (High Sales Threshold - purchased_last_month): {high_sales_threshold:.0f}")

# Calculate the 25th percentile of 'purchased_last_month' to define the low sales threshold.
low_sales_threshold = kpi_df['purchased_last_month'].quantile(0.25)
print(f"25th percentile (Low Sales Threshold - purchased_last_month): {low_sales_threshold:.0f}")

# 2. Identify products with High Ratings but Low Sales:
high_rating_low_sales_df = kpi_df[
    (kpi_df['product_rating'] >= high_rating_threshold) &
    (kpi_df['purchased_last_month'] <= low_sales_threshold)
].copy()

# 3. Identify products with Low Ratings but High Sales:
low_rating_high_sales_df = kpi_df[
    (kpi_df['product_rating'] <= low_rating_threshold) &
    (kpi_df['purchased_last_month'] >= high_sales_threshold)
].copy()

# 4. Display the count of products in each DataFrame.
print(f"\nNumber of products with High Ratings but Low Sales: {len(high_rating_low_sales_df)}")
print(f"Number of products with Low Ratings but High Sales: {len(low_rating_high_sales_df)}")

# 5. Print the first 5 rows of each DataFrame with specified columns.
columns_to_display = ['product_title', 'product_category', 'product_rating', 'purchased_last_month', 'discounted_price', 'actual_revenue']

print("\n--- High Ratings but Low Sales Products ---")
if not high_rating_low_sales_df.empty:
    print(high_rating_low_sales_df[columns_to_display].head())
else:
    print("No products found in this category.")

print("\n--- Low Ratings but High Sales Products ---")
if not low_rating_high_sales_df.empty:
    print(low_rating_high_sales_df[columns_to_display].head())
else:
    print("No products found in this category.")
```

## Assess Discount Addiction

### Subtask:
Determine if customers are primarily buying products only when discounts exceed a certain threshold (e.g., 40%). This will involve categorizing sales by discount percentage and analyzing purchasing behavior.


**Reasoning**:
The subtask requires defining a discount threshold, categorizing products based on this threshold into highly discounted and lower discounted sales, calculating total purchased last month and actual revenue for each category, and then computing their percentage contributions to overall sales and revenue. These steps will be performed in a single code block.



```python
import numpy as np

# 1. Define a discount threshold
discount_threshold_value = 40.0
print(f"Discount Threshold defined: {discount_threshold_value:.0f}%")

# 2. Filter the kpi_df DataFrame
highly_discounted_sales_df = kpi_df[kpi_df['discount_percentage'] >= discount_threshold_value].copy()
lower_discount_sales_df = kpi_df[kpi_df['discount_percentage'] < discount_threshold_value].copy()

# 3. Calculate total 'purchased_last_month' and 'actual_revenue' for each DataFrame
total_purchased_highly_discounted = highly_discounted_sales_df['purchased_last_month'].sum()
total_revenue_highly_discounted = highly_discounted_sales_df['actual_revenue'].sum()

total_purchased_lower_discount = lower_discount_sales_df['purchased_last_month'].sum()
total_revenue_lower_discount = lower_discount_sales_df['actual_revenue'].sum()

# 4. Calculate overall total 'purchased_last_month' and 'actual_revenue' for all products in kpi_df
total_purchased_overall = kpi_df['purchased_last_month'].sum()
total_revenue_overall = kpi_df['actual_revenue'].sum()

# 5. Compute the percentage contribution
percentage_purchased_highly_discounted = (total_purchased_highly_discounted / total_purchased_overall) * 100 if total_purchased_overall > 0 else 0
percentage_revenue_highly_discounted = (total_revenue_highly_discounted / total_revenue_overall) * 100 if total_revenue_overall > 0 else 0

percentage_purchased_lower_discount = (total_purchased_lower_discount / total_purchased_overall) * 100 if total_purchased_overall > 0 else 0
percentage_revenue_lower_discount = (total_revenue_lower_discount / total_revenue_overall) * 100 if total_revenue_overall > 0 else 0

# 6. Print the results
print("\n--- Analysis of Discount Addiction (Threshold > 40%) ---")

print("\nHIGHLY DISCOUNTED PRODUCTS (Discount >= 40%):")
print(f"  Number of products: {len(highly_discounted_sales_df)}")
print(f"  Total Purchased Last Month: {total_purchased_highly_discounted:,.0f}")
print(f"  Total Actual Revenue: ${total_revenue_highly_discounted:,.2f}")
print(f"  % Contribution to Overall Purchased Last Month: {percentage_purchased_highly_discounted:.2f}%")
print(f"  % Contribution to Overall Actual Revenue: {percentage_revenue_highly_discounted:.2f}%")

print("\nLOWER DISCOUNT PRODUCTS (Discount < 40%):")
print(f"  Number of products: {len(lower_discount_sales_df)}")
print(f"  Total Purchased Last Month: {total_purchased_lower_discount:,.0f}")
print(f"  Total Actual Revenue: ${total_revenue_lower_discount:,.2f}")
print(f"  % Contribution to Overall Purchased Last Month: {percentage_purchased_lower_discount:.2f}%")
print(f"  % Contribution to Overall Actual Revenue: {percentage_revenue_lower_discount:.2f}%")

print("\nOVERALL TOTALS:")
print(f"  Total Purchased Last Month (Overall): {total_purchased_overall:,.0f}")
print(f"  Total Actual Revenue (Overall): ${total_revenue_overall:,.2f}")
```

## Identify Category Subsidies

### Subtask:
Pinpoint which product categories are heavily reliant on discounts to drive sales, essentially being 'funded' by discounts rather than inherent value.


**Reasoning**:
I need to calculate the discount reliance percentage for each product category using the `revenue_per_category` DataFrame, which involves computing the absolute difference between potential and actual revenue and then dividing by potential revenue. I will then sort the results and display the top 10 categories.



```python
import numpy as np

# 1. Calculate the absolute difference between 'Potential Revenue without Discounts per Category' and 'Revenue with Discounts per Category'
# to determine the total discount amount for each category.
revenue_per_category['discount_amount_per_category'] = revenue_per_category['Potential Revenue without Discounts per Category'] - revenue_per_category['Revenue with Discounts per Category']

# 2. Calculate the 'Discount Reliance Percentage' for each category.
# This can be done by dividing the total discount amount by the 'Potential Revenue without Discounts per Category' and multiplying by 100.
# Handle potential division by zero if 'Potential Revenue without Discounts per Category' is 0.
revenue_per_category['Discount Reliance Percentage'] = (
    revenue_per_category['discount_amount_per_category'] / revenue_per_category['Potential Revenue without Discounts per Category']
) * 100
revenue_per_category['Discount Reliance Percentage'].fillna(0, inplace=True) # Fill NaN if potential revenue was 0

# 3. Sort the categories in descending order based on their 'Discount Reliance Percentage' to highlight categories most reliant on discounts.
revenue_per_category_sorted = revenue_per_category.sort_values(by='Discount Reliance Percentage', ascending=False)

# 4. Display the top 10 categories with the highest discount reliance, showing their 'product_category',
# 'Revenue with Discounts per Category', 'Potential Revenue without Discounts per Category', and the calculated 'Discount Reliance Percentage'.
print("\n--- Top 10 Product Categories Heavily Reliant on Discounts ---")
print(revenue_per_category_sorted[[
    'product_category',
    'Revenue with Discounts per Category',
    'Potential Revenue without Discounts per Category',
    'Discount Reliance Percentage'
]].head(10))
```

**Reasoning**:
The previous code produced a `FutureWarning` because of the `inplace=True` argument used on a chained assignment for `fillna`. To fix this, I will assign the result of the `fillna` operation directly back to the column, which is the recommended practice to avoid such warnings and ensure the DataFrame is modified correctly.



```python
import numpy as np

# 1. Calculate the absolute difference between 'Potential Revenue without Discounts per Category' and 'Revenue with Discounts per Category'
# to determine the total discount amount for each category.
revenue_per_category['discount_amount_per_category'] = revenue_per_category['Potential Revenue without Discounts per Category'] - revenue_per_category['Revenue with Discounts per Category']

# 2. Calculate the 'Discount Reliance Percentage' for each category.
# This can be done by dividing the total discount amount by the 'Potential Revenue without Discounts per Category' and multiplying by 100.
# Handle potential division by zero if 'Potential Revenue without Discounts per Category' is 0.
revenue_per_category['Discount Reliance Percentage'] = (
    revenue_per_category['discount_amount_per_category'] / revenue_per_category['Potential Revenue without Discounts per Category']
) * 100
# Fix: Assign the result of fillna directly to the column to avoid FutureWarning
revenue_per_category['Discount Reliance Percentage'] = revenue_per_category['Discount Reliance Percentage'].fillna(0)

# 3. Sort the categories in descending order based on their 'Discount Reliance Percentage' to highlight categories most reliant on discounts.
revenue_per_category_sorted = revenue_per_category.sort_values(by='Discount Reliance Percentage', ascending=False)

# 4. Display the top 10 categories with the highest discount reliance, showing their 'product_category',
# 'Revenue with Discounts per Category', 'Potential Revenue without Discounts per Category', and the calculated 'Discount Reliance Percentage'.
print("\n--- Top 10 Product Categories Heavily Reliant on Discounts ---")
print(revenue_per_category_sorted[[
    'product_category',
    'Revenue with Discounts per Category',
    'Potential Revenue without Discounts per Category',
    'Discount Reliance Percentage'
]].head(10))
```

Examine how time-to-delivery affects customer retention or repeat purchases within different categories. This will require customer and order history data to track repeat behavior.

I need to calculate the delivery duration, handle potential negative values, calculate the overall correlation between delivery duration and sales, and then perform a category-wise analysis of average delivery duration and total sales as per the subtask instructions. All these steps will be performed in a single code block.



```python
import numpy as np

# 1. Calculate delivery_duration_days
# Assuming data_collected_at as order_date for this analysis
kpi_df['delivery_duration'] = kpi_df['delivery_date'] - kpi_df['data_collected_at']
kpi_df['delivery_duration_days'] = kpi_df['delivery_duration'].dt.days

# Handle potential negative durations: convert them to NaN (as they indicate anomalies)
# Or, set to 0 if 'data_collected_at' is after 'delivery_date' (meaning instantaneous or same-day delivery conceptually, or error)
# For correlation analysis, NaN is generally better as it's excluded, but for overall average, 0 might be more appropriate if it means same day. Let's use NaN to avoid biasing the average with potentially erroneous 0s.
kpi_df.loc[kpi_df['delivery_duration_days'] < 0, 'delivery_duration_days'] = np.nan

print("Delivery duration calculated and negative durations handled.")

# 2. Calculate the Pearson correlation coefficient between delivery_duration_days and purchased_last_month
delivery_sales_correlation = kpi_df['delivery_duration_days'].corr(kpi_df['purchased_last_month'])

# 3. Group by product_category and calculate mean delivery_duration_days and sum of purchased_last_month
category_delivery_sales = kpi_df.groupby('product_category').agg(
    avg_delivery_duration_days=('delivery_duration_days', 'mean'),
    total_purchased_last_month=('purchased_last_month', 'sum')
).reset_index()

# 4. Print the overall correlation
print(f"\nOverall Pearson Correlation between Delivery Duration (days) and Purchased Last Month: {delivery_sales_correlation:.2f}")

# 5. Display category-wise average delivery duration and total sales, sorted by average delivery duration
category_delivery_sales_sorted = category_delivery_sales.sort_values(by='avg_delivery_duration_days', ascending=False)
print("\nCategory-wise Average Delivery Duration and Total Sales (Sorted by Average Delivery Duration):")
print(category_delivery_sales_sorted.head(10)) # Display top 10 categories with longest average delivery
```
